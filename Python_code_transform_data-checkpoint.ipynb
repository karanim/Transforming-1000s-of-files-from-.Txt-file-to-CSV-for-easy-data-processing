```python
import os
import pandas as pd
import chardet

def detect_encoding(file_path):
    with open(file_path, 'rb') as f:
        result = chardet.detect(f.read())
    return result['encoding']

def txt_to_xlsx(txt_filepath, xlsx_filepath, delimiter='\t'):
    encoding = detect_encoding(txt_filepath)
    df = pd.read_csv(txt_filepath, sep=delimiter, engine='python', encoding=encoding)
    df.to_excel(xlsx_filepath, index=False)

def batch_convert_txt_to_xlsx(txt_folder, xlsx_folder):
    if not os.path.exists(xlsx_folder):
        os.makedirs(xlsx_folder)

    txt_files = [f for f in os.listdir(txt_folder) if f.endswith('.txt')]

    for txt_file in txt_files:
        txt_filepath = os.path.join(txt_folder, txt_file)
        xlsx_filepath = os.path.join(xlsx_folder, os.path.splitext(txt_file)[0] + '.xlsx')
        try:
            txt_to_xlsx(txt_filepath, xlsx_filepath)
            print(f"Conversion successful: {txt_filepath} to {xlsx_filepath}")
        except Exception as e:
            print(f"Error converting {txt_filepath}: {e}")

# Replace 'Losartan_files' and 'Losartan_xlsx_files' with your actual folder names
txt_folder = r'C:\Users\Wallstreet consult\Desktop\Metronidazole TXT'
xlsx_folder = 'Metronidazole_xlsx'

batch_convert_txt_to_xlsx(txt_folder, xlsx_folder)
```
